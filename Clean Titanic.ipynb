### Cleaning

#### Column Numbers

```python
# Column Numbers
c1 = "passengerid"
c2 = "pclass"
c3 = "name"
c4 = "sex"
c5 = "age"
c6 = "sibsp"
c7 = "parch"
c8 = "ticket"
c9 = "fare"
c10 = "cabin"
c11 = "embarked"
```

#### Null Values

```python
# Null Values
null_values=["na","n/a","none","null","","nan","nat","UNKNOWN","ERROR"]
```

#### Column Values

```python
# Column Values
df[c3] = df[c3].astype(str).apply(lambda x: ' '.join([part.strip() for part in x.split(',')[::-1]]))
df[c3] = df[c3].str.extract(r'(?:Mr\.|Mrs\.)\s*([^,()]+)')
df[c4] = df[c4].str.lower()
df[c4] = df[c4].replace({
    'male': 'M',
    'female': 'F'
})
```

#### Missing

```python
# Triming
for col in df.columns:
    clean = df[col].astype(str).str.strip()
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = pd.to_numeric(clean, errors='coerce')
    elif pd.api.types.is_datetime64_any_dtype(df[col]):
        df[col] = pd.to_datetime(clean, errors='coerce')
    elif pd.api.types.is_bool_dtype(df[col]):
        df[col] = clean.map({'True': True, 'False': False})
    elif pd.api.types.is_categorical_dtype(df[col]):
        df[col] = clean.astype('category')
    else:
        df[col] = clean

# Null Convert
df = df.replace(to_replace=null_values, value=np.nan)

# Missing Values
for col in df.columns:
    if df[col].isnull().any():
        df[col + '_isnull'] = (~df[col].isnull()).astype(int)

# Datetime
dates = [i for i in df.columns if any(j in i for j in ["date","dt","timestamp"])]
times = [i for i in df.columns if any(j in i for j in ["time"])]
for col in dates:
    df[col] = pd.to_datetime(df[col], errors='coerce')
for col in times:
    temp = pd.to_datetime(df[col], format='%H:%M:%S', errors='coerce')
    df[col] = pd.to_datetime('1970-01-01') + (temp - temp.dt.normalize())
```

#### Values

```python
# Values

df[c10] = df[c10].notna().astype(int)
df=df.sort_values(by=[f'{c4}',f'{c5}'], ascending=[True,False])
limits = {
    f'{c5}': (1, 100)
}
for col, (low, high) in limits.items():
    df.loc[~df[col].between(low, high), col] = None
    df[col] = df[col].fillna(df[col].median())


cols_to_clean = [c8]
for col in cols_to_clean:
    df[col] = df[col].str.extract(r'(\d+)')[0]
    df = df.dropna(subset=[col])
cols_to_convert = [f"{c5}"]
df[cols_to_convert] = df[cols_to_convert].astype("int64")
df = df.round({
    f'{c9}': 2
})
```
